<p style='text-align:justified'><b>1.</b> Aplique la operación de dilatación sobre la imagen siguiente con distintos elementos estructurantes (SE) y observe las diferencias en los resultados.

```python
# Se importa la imagen
img = cv2.imread('figs/dilationt1.png',0)
# Se le aplica umbralización binaria invertida para ser operada por el SE
ret,img_th = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
# Se crean los elementos estructurantes
se1 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
se2 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
se3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,5))
# Se aplican las operaciones de dilatación
img_dil1 = cv2.dilate(img_th,se1)
img_dil2 = cv2.dilate(img_th,se2)
img_dil3 = cv2.dilate(img_th,se3)
# Se muestran los resultados
plt.figure(1)
plt.subplots_adjust(top=2,bottom=0.01,left=0.10,right=2,hspace=0.1,wspace=0.1)
plt.subplot(431),plt.imshow(img,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(434),plt.imshow(se1,cmap='gray')
plt.title('SE - Rectangular'),plt.xticks([]),plt.yticks([])
plt.subplot(435),plt.imshow(se2,cmap='gray')
plt.title('SE - Cross'),plt.xticks([]),plt.yticks([])
plt.subplot(436),plt.imshow(se3,cmap='gray')
plt.title('SE - Ellipse'),plt.xticks([]),plt.yticks([])
plt.subplot(437),plt.imshow(img_dil1,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(438),plt.imshow(img_dil2,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(439),plt.imshow(img_dil3,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,10),plt.imshow(img_dil1[114:155, 40:70],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,11),plt.imshow(img_dil2[114:155, 40:70],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,12),plt.imshow(img_dil3[114:155, 40:70],cmap='gray')
plt.xticks([]),plt.yticks([])
```

<p>Un análisis general del texto y uno más específico, analizando una de las letras, presentado en la tercera y cuarta filas de imágenes respectivamente, permite ver un engrosamiento en cada una de las letras del mismo, pero con diferenciaciones algo sutiles; la primera columna, muestra la aplicación del SE tipo rectangular, el cual posee en ensanchamiento general en todas direcciones pero de una forma en que cada pixel de la letra se rellena de manera recta, es decir, generando superficies horizontales y verticales cada vez más lisas; en la segunda columna es posible visualizar un efecto algo distinto, aquí en ensanchamiento no es tan marcado y más bien pareciera visualizar un patrón de tipo escalera entre los píxeles; por último, en la tercera columna si existe una relación más cercana con el caso anterior, a nivel general el resultado es muy parecido pero, cuando se mira más de cerca, se alcanza a notar un grado de ensanchamiento que se favorece en el eje vertical y que esta enteramente dado por como se definió el SE.</p>
<p>En general se puede apreciar que cada vez más las letras tienden a tomar la forma del elemento estructurante.</p>

<p style='text-align:justified'><b>2.</b> Aplique la operación de erosión sobre la imagen anterior (sin dilatar). ¿Qué observa?</p>

```python
# Se importa la imagen
img = cv2.imread('figs/dilationt1.png',0)
# Se le aplica umbralización binaria invertida para ser operada por el SE
ret,img_th = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
# Se crean los elementos estructurantes
se1 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
se2 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
se3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,5))
# Se aplican las operaciones de dilatación
img_ero1 = cv2.erode(img_th,se1)
img_ero2 = cv2.erode(img_th,se2)
img_ero3 = cv2.erode(img_th,se3)
# Se muestran los resultados
plt.figure(1)
plt.subplots_adjust(top=2,bottom=0.01,left=0.10,right=2,hspace=0.1,wspace=0.1)
plt.subplot(431),plt.imshow(img,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(435),plt.imshow(se1,cmap='gray')
plt.title('SE - Rectangular'),plt.xticks([]),plt.yticks([])
plt.subplot(434),plt.imshow(se2,cmap='gray')
plt.title('SE - Cross'),plt.xticks([]),plt.yticks([])
plt.subplot(436),plt.imshow(se3,cmap='gray')
plt.title('SE - Ellipse'),plt.xticks([]),plt.yticks([])
plt.subplot(438),plt.imshow(img_ero1,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(437),plt.imshow(img_ero2,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(439),plt.imshow(img_ero3,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,11),plt.imshow(img_ero1[114:155, 40:70],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,10),plt.imshow(img_ero2[114:155, 40:70],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,12),plt.imshow(img_ero3[114:155, 40:70],cmap='gray')
plt.xticks([]),plt.yticks([])
```

<p>La operación de erosión, a diferencia de la de dilatación, conserva los píxeles que cumplan la condición en donde el SE este contenido dentro del conjunto de píxeles cuando este se ubica de manera coincidente con el anchor y el píxel a analizar. De esta manera se puede decir que la condición es mucho más restrictiva y que depende del tamaño del SE, si este es muy grande, existirán cada vez menos conjunto de píxeles de cumplan con dicho parámetro. Es por ello que, si se observan los resultados de las filas tres y cuatro, de izquierda a derecha, los resultados tienden a mostrar cada vez menos partes del texto, esto es debido a que se aplican muchas más condiciones a los conjuntos de píxeles analizados y, de hecho, si se mira el análisis por letra, la cantidad de píxeles mostrados disminuye haciendo a la letra cada vez menos distinguible.</p>

<p style='text-align:justified'><b>3.</b> Tome una imagen de su elección y umbralícela para obtenerla a blanco y negro. Luego:<br><br>
a) Aplique una dilatación y luego una erosión sobre ella, y verifique si la erosión deshace los cambios realizados por la dilatación.</p>

```python
# Se obtiene la imagen del dataset de imagenes obtenido para realizar el proyecto de curso
img = cv2.imread('figs/62.png',0)[180:400,250:400]
# Se umbraliza la imagen
ret,thresh1 = cv2.threshold(img,90,255,cv2.THRESH_BINARY_INV)
# Se obtiene el SE
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
# Se realiza el proceso de dilatación
img_dil1 = cv2.dilate(thresh1,se1)
# Se realiza el proceso de erosión consecutivo
img_er1 = cv2.erode(img_dil1,se1)

# Se muestran los resultados
plt.subplots(figsize=(15,15))
plt.subplots_adjust(top=1,bottom=0.2,left=0.2,right=1,hspace=0.2,wspace=0.5)
plt.subplot(431),plt.imshow(img,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(434),plt.imshow(thresh1,cmap='gray')
plt.title('Umbralizada'),plt.xticks([]),plt.yticks([])
plt.subplot(435),plt.imshow(img_dil1,cmap='gray')
plt.title('Dilatada'),plt.xticks([]),plt.yticks([])
plt.subplot(436),plt.imshow(img_er1,cmap='gray')
plt.title('Erosionada'),plt.xticks([]),plt.yticks([])
plt.subplot(437),plt.imshow(thresh1[85:100,85:100],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(438),plt.imshow(img_dil1[85:100,85:100],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(439),plt.imshow(img_er1[85:100,85:100],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,10),plt.imshow(thresh1[5:20,135:150],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,11),plt.imshow(img_dil1[5:20,135:150],cmap='gray')
plt.xticks([]),plt.yticks([])
plt.subplot(4,3,12),plt.imshow(img_er1[5:20,135:150],cmap='gray')
plt.xticks([]),plt.yticks([])
```

<p>En efecto una operación de erosión seguida por una dilatación, utilizando el mismo SE, permite deshacer los efectos ocasionados por la función de dilatación. Analizando las imágenes y una pequeña parte de estas a nivel de pixeles en dos zonas distintas, es posible observar que esta secuencia de operaciones permite regresar la imagen a su estado original ya sea en una porción central o marginal.</p>
<br>
b) Dilátela usando el mismo SE del inciso anterior, y luego halle el complemento del resultado que obtenga (lo que era blanco, ahora será negro y viceversa).

```python
# Se obtiene la imagen del dataset de imagenes obtenido para realizar el proyecto de curso
img = cv2.imread('figs/62.png',0)[180:400,250:400]
# Se umbraliza la imagen
ret,thresh1 = cv2.threshold(img,90,255,cv2.THRESH_BINARY_INV)
# Se obtiene el SE
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
# Se realiza el proceso de erosión
img_dil1 = cv2.dilate(thresh1,se1)
# Se obtiene el complemento de la imagen
img_comp = 255 - img_dil1


# Se muestran los resultados
plt.subplots(figsize=(15,15))
plt.subplots_adjust(top=1,bottom=0.2,left=0.2,right=1,hspace=0.2,wspace=0.5)
plt.subplot(121),plt.imshow(img_dil1,cmap='gray')
plt.title('Dilatación'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(img_comp,cmap='gray')
plt.title('Complemento'),plt.xticks([]),plt.yticks([])
```

c) Compare el resultado del inciso b) con el obtenido al aplicar la operación $A^c \ominus \hat{B}$, donde $A$ es su imagen en formato blanco y negro (use el mismo SE del inciso a). Esto se conoce como el <b> principio de dualidad</b>: La dilatación y la erosión no son opuestas, sino que se pueden escribir una en función de la otra.

```python
# Se obtiene la imagen del dataset de imagenes obtenido para realizar el proyecto de curso
img = cv2.imread('figs/62.png',0)[180:400,250:400]
# Se umbraliza la imagen
ret,thresh1 = cv2.threshold(img,90,255,cv2.THRESH_BINARY_INV)
# Se obtiene el complemento de la imagen
img_comp = 255 - thresh1
# Se obtiene el SE
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
# Se realiza el proceso de erosión
img_ero1 = cv2.erode(img_comp,se1)

# Se muestran los resultados
plt.subplots(figsize=(15,15))
plt.subplots_adjust(top=1,bottom=0.2,left=0.2,right=1,hspace=0.2,wspace=0.5)
plt.subplot(121),plt.imshow(img_comp,cmap='gray')
plt.title('Complemento'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(img_ero1,cmap='gray')
plt.title('Erosión'),plt.xticks([]),plt.yticks([])
```

Se puede observar que de los incisos b) y c) llegan al mismo resultado, a pesar de que las operaciones utilizadas sean distintas. Se puede entonces comprobar la afirmación hecha en el el enciso c) "La dilatación y la erosión no son opuestas, sino que se pueden escribir una en función de la otra".

El SE tipo cruz no mostró un desempeño particularmente optimo sobre ninguna de las formas presentes en la figura.</p>

<p>En general, se puede decir que las operaciones de apertura tienen un mejor desempeño en eliminar el ruido tipo sal, mientras que las operaciones de tipo cierre tienen un mejor desempeño tratando con el ruido tipo pimienta.</p>

```python
# Se obtiene la imagen del dataset de imagenes obtenido para realizar el proyecto de curso
img = cv2.imread('figs/62.png',0)[180:400,250:400]
# Se implementan las operaciones de suavizado de contornos
sal1 = cv2.blur(img, (5, 5))
sal2 = cv2.medianBlur(img, 5)
sal3 = cv2.GaussianBlur(img, (5, 5), 0)
sal4 = cv2.bilateralFilter(img, 5, 75, 75)
# Se aplican la operaciones de apertura y cierre para generar el efecto de suavizado
se = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
img_op = cv2.morphologyEx(img, cv2.MORPH_OPEN, se)
img_cl = cv2.morphologyEx(img_op, cv2.MORPH_CLOSE, se)

# Se muestran los resultados
plt.subplots(figsize=(15,15))
plt.subplots_adjust(top=1,bottom=0.2,left=0.2,right=1,hspace=0.2,wspace=0.5)
plt.subplot(231), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(232), plt.imshow(sal1, cmap='gray')
plt.title('Blur'), plt.xticks([]), plt.yticks([])
plt.subplot(233), plt.imshow(sal2, cmap='gray')
plt.title('Median'), plt.xticks([]), plt.yticks([])
plt.subplot(234), plt.imshow(sal3, cmap='gray')
plt.title('Gaussian'), plt.xticks([]), plt.yticks([])
plt.subplot(235), plt.imshow(sal4, cmap='gray')
plt.title('Bilateral'), plt.xticks([]), plt.yticks([])
plt.subplot(236), plt.imshow(img_cl, cmap='gray')
plt.title('Apertura-Cierre'), plt.xticks([]), plt.yticks([])
plt.show()
```

<p>Se puede observar que existen diferencias significativas entre cada una de las técnicas de suavizado pero, evidentemente, la de apertura y cierre tiene un efecto más agresivo sobre la imagen, a tal punto de caricaturizar la imagen e incluso de eliminar aspectos de la misma como se puede apreciar en la parte superior derecha, dando como resultado una simplificación de la imagen en sus detalles, colores y formas.</p>
<p style='text-align:justify'><b>7.</b> Ejecute el script revisado arriba para la transformación Hit-or-Miss, pero esta vez utilice como SE la letra <i>t</i> (la ruta es "figs/sehom3.png"). ¿Se reconocen correctamente todas las coincidencias? En caso de que no, ¿Qué se puede hacer para corregir esto? </p>

```python
# Se obtiene la imagen de texto
img = cv2.imread('figs/dilationt1.png',0)
# Se umbraliza la imagen y se invierte su reultado para que sea operabla
ret,img_th = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
#Se obtiene el SE de la letra t
sel = cv2.imread('figs/sehom3.png',0)

# Se modifica el SE en caso de que sus dimensiones en x o en y sean pares (el SE debe ser impar en todas sus dimensiones)
if (sel.shape[0]%2) == 0:
    sel = sel[:sel.shape[0]-1,:]
if (sel.shape[1]%2) == 0:
    sel = sel[:,:sel.shape[1]-1] 
# Se umbraliza a formato binario el SE y se invierten sus valores para que coincida
ret,sel = cv2.threshold(sel,127,255,cv2.THRESH_BINARY_INV)
# El SE debe solo contener valores entre 0 y 1 por lo que se debe normalizar  
se = sel//255

se_aux = np.asarray([[1],[1]])
se = cv2.erode(se,se_aux)        #SE is eroded to allow a more flexible match

# Se aplica el algorítmo Hit or Miss
img_hom = cv2.morphologyEx(img_th,cv2.MORPH_HITMISS,se)
# Esta seccion de código imprime el elemento estructurante original en el punto de coincidencia ubicado por 
# el algorítmo Hit or Miss  
img_res = np.zeros(img.shape)    #Array to store processed image
h = se.shape[0]//2               #Half the height of SE
w = se.shape[1]//2               #Half the width of SE
counter = 0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img_hom[i,j] == 255:
            if j>=w:             # fixes matching bugs at image borders
                img_res[i-h:i+h+1,j-w:j+w+1] = sel
                counter += 1
img_res = 255-img_res

print(f"Se han detectado {counter} coincidencias.")
 
# Se muestran los resultados
plt.subplots(figsize=(15,15))
plt.subplots_adjust(top=0.5,bottom=0.2,left=0.2,right=1,hspace=0.1,wspace=0.5)
plt.subplot(221),plt.title('Letra t'),plt.imshow(255-sel,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.title('SE'),plt.imshow(255-se,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(223),plt.title('Original'),plt.imshow(img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(224),plt.title('Hit-or-Miss'),plt.imshow(img_res,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
```

De manera inicial, utilizando el SE tal cual es proporcionado no arroja ninguna coincidencia. Para solucionar este problema se recurre a erosionar el SE para hacerlo más delgado y que pueda tener un mayor grado de relacionamiento a ser este más permisivo en su condicionamiento. Con este arreglo fue posible determinar la cantidad de "t"s presentes en el texto dando como resultado un total de 12 conteos.
