```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import skimage.morphology as morph
```

```python
img1 = cv.imread("human detection dataset/1/111.png",0)
img2 = cv.imread("human detection dataset/1/205.png",0)

plt.figure(figsize=(10,5))
plt.subplot(121); plt.imshow(img1, cmap='gray'); plt.title("Imagen 1")
plt.subplot(122); plt.imshow(img2, cmap='gray'); plt.title("Imagen 2")
```

# Dilatación

```python
ret, img1_th = cv.threshold(img1, 50, 255, cv.THRESH_BINARY_INV)
ret, img2_th = cv.threshold(img2, 80, 255, cv.THRESH_BINARY_INV)

#Dilatación
se1 = cv.getStructuringElement(cv.MORPH_CROSS, (5,5))
se2 = cv.getStructuringElement(cv.MORPH_RECT,(3,5))

img1_dil = cv.dilate(img1_th, se1, iterations = 1)
img2_dil = cv.dilate(img2_th, se2, iterations = 1)

plt.figure(figsize=(10,5))
plt.subplot(231); plt.imshow(img1_th, cmap='gray'); plt.title("Imagen 1"); plt.axis("off")
plt.subplot(234); plt.imshow(img2_th, cmap='gray'); plt.title("Imagen 2"); plt.axis("off")
plt.subplot(232); plt.imshow(se1, cmap='gray'); plt.title("SE - CROSS"); plt.axis("off")
plt.subplot(235); plt.imshow(se2, cmap='gray'); plt.title("SE - RECT"); plt.axis("off")
plt.subplot(233); plt.imshow(img1_dil, cmap='gray'); plt.title("Imagen 1 | SE - CROSS"); plt.axis("off")
plt.subplot(236); plt.imshow(img2_dil, cmap='gray'); plt.title("Imagen 2 | SE - RECT"); plt.axis("off")
```

# Erosión

```python
ret, img1_th = cv.threshold(img1, 50, 255, cv.THRESH_BINARY_INV)
ret, img2_th = cv.threshold(img2, 80, 255, cv.THRESH_BINARY_INV)

#Erosion
se1 = cv.getStructuringElement(cv.MORPH_CROSS, (5,5))
se2 = cv.getStructuringElement(cv.MORPH_RECT,(3,5))

img1_er = cv.erode(img1_th, se1, iterations = 1)
img2_er = cv.erode(img2_th, se2, iterations = 1)

plt.figure(figsize=(10,5))
plt.subplot(231); plt.imshow(img1_th, cmap='gray'); plt.title("Imagen 1"); plt.axis("off")
plt.subplot(234); plt.imshow(img2_th, cmap='gray'); plt.title("Imagen 2"); plt.axis("off")
plt.subplot(232); plt.imshow(se1, cmap='gray'); plt.title("SE - CROSS"); plt.axis("off")
plt.subplot(235); plt.imshow(se2, cmap='gray'); plt.title("SE - RECT"); plt.axis("off")
plt.subplot(233); plt.imshow(img1_er, cmap='gray'); plt.title("Imagen 1 | SE - CROSS"); plt.axis("off")
plt.subplot(236); plt.imshow(img2_er, cmap='gray'); plt.title("Imagen 2 | SE - RECT"); plt.axis("off")
```

# Apertura


```python
ret, img1_th = cv.threshold(img1, 50, 255, cv.THRESH_BINARY_INV)
ret, img2_th = cv.threshold(img2, 80, 255, cv.THRESH_BINARY_INV)

#Apertura
se1 = cv.getStructuringElement(cv.MORPH_CROSS, (7,7))
se2 = cv.getStructuringElement(cv.MORPH_RECT,(3,5))

img1_ap = cv.morphologyEx(img1_th, cv.MORPH_OPEN, se1)
img2_ap = cv.morphologyEx(img2_th, cv.MORPH_OPEN, se2) 

plt.figure(figsize=(10,5))
plt.subplot(231); plt.imshow(img1_th, cmap='gray'); plt.title("Imagen 1"); plt.axis("off")
plt.subplot(234); plt.imshow(img2_th, cmap='gray'); plt.title("Imagen 2"); plt.axis("off")
plt.subplot(232); plt.imshow(se1, cmap='gray'); plt.title("SE - CROSS"); plt.axis("off")
plt.subplot(235); plt.imshow(se2, cmap='gray'); plt.title("SE - RECT"); plt.axis("off")
plt.subplot(233); plt.imshow(img1_ap, cmap='gray'); plt.title("Imagen 1 | SE - CROSS"); plt.axis("off")
plt.subplot(236); plt.imshow(img2_ap, cmap='gray'); plt.title("Imagen 2 | SE - RECT"); plt.axis("off")
```

# Cierre

```python
ret, img1_th = cv.threshold(img1, 50, 255, cv.THRESH_BINARY_INV)
ret, img2_th = cv.threshold(img2, 80, 255, cv.THRESH_BINARY_INV)

#Cierre
se1 = cv.getStructuringElement(cv.MORPH_CROSS, (7,7))
se2 = cv.getStructuringElement(cv.MORPH_RECT,(3,5))

img1_cl = cv.morphologyEx(img1_th, cv.MORPH_CLOSE, se1)
img2_cl = cv.morphologyEx(img2_th, cv.MORPH_CLOSE, se2) 

plt.figure(figsize=(10,5))
plt.subplot(231); plt.imshow(img1_th, cmap='gray'); plt.title("Imagen 1"); plt.axis("off")
plt.subplot(234); plt.imshow(img2_th, cmap='gray'); plt.title("Imagen 2"); plt.axis("off")
plt.subplot(232); plt.imshow(se1, cmap='gray'); plt.title("SE - CROSS"); plt.axis("off")
plt.subplot(235); plt.imshow(se2, cmap='gray'); plt.title("SE - RECT"); plt.axis("off")
plt.subplot(233); plt.imshow(img1_cl, cmap='gray'); plt.title("Imagen 1 | SE - CROSS"); plt.axis("off")
plt.subplot(236); plt.imshow(img2_cl, cmap='gray'); plt.title("Imagen 2 | SE - RECT"); plt.axis("off")
```

# Transformada Hit-or-Miss

```python
sel1 = img1[15:220, 155:220]
sel2 = img2[20:140, 115:170]

# ajustamos el SE para que sea impar
if (sel1.shape[0]%2) == 0:
    sel1 = sel1[:sel1.shape[0]-1, :]
if (sel1.shape[1]%2) == 0:
    sel1 = sel1[:,:sel1.shape[1]-1]
    
if (sel2.shape[0]%2) == 0:
    sel2 = sel2[:sel2.shape[0]-1, :]
if (sel2.shape[1]%2) == 0:
    sel2 = sel2[:,:sel2.shape[1]-1]
    
ret,sel1 = cv.threshold(sel1, 50, 255, cv.THRESH_BINARY_INV)
ret,sel2 = cv.threshold(sel2, 80, 255, cv.THRESH_BINARY_INV)

se1 = sel1//255
se2 = sel2//255



img1_hom = cv.morphologyEx(img1_th, cv.MORPH_HITMISS, se1)
img2_hom = cv.morphologyEx(img2_th, cv.MORPH_HITMISS, se2)
```

```python
def get_res_img(img, img_hom, se, sel):
    
    img_res = np.zeros(img.shape)    
    h = se.shape[0]//2              
    w = se.shape[1]//2               

    counter = 0
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img_hom[i,j] == 255:
                if j>=w:             
                    img_res[i-h:i+h+1,j-w:j+w+1] = sel
                    counter += 1
    img_res = 255-img_res
    
    return img_res
```

```python
img1_res = get_res_img(img1, img1_hom, se1, sel1)
img2_res = get_res_img(img2, img2_hom, se2, sel2)

plt.figure(figsize=(10,5))
plt.subplot(231); plt.imshow(img1_th, cmap='gray'); plt.title("Imagen 1"); plt.axis("off")
plt.subplot(234); plt.imshow(img2_th, cmap='gray'); plt.title("Imagen 2"); plt.axis("off")
plt.subplot(232); plt.imshow(se1, cmap='gray'); plt.title("SE2"); plt.axis("off")
plt.subplot(235); plt.imshow(se2, cmap='gray'); plt.title("SE2"); plt.axis("off")
plt.subplot(233); plt.imshow(img1_res, cmap='gray'); plt.title("Imagen 1 | Coincidencias"); plt.axis("off")
plt.subplot(236); plt.imshow(img2_res, cmap='gray'); plt.title("Imagen 2 | Coincidencias"); plt.axis("off")
```

# Llenado de regiones

```python
ret, img1_th = cv.threshold(img1, 50, 255, cv.THRESH_BINARY_INV)
ret, img2_th = cv.threshold(img2, 80, 255, cv.THRESH_BINARY_INV)


#Boundary closing through dilation
se = np.asarray([[1,1,1]], np.uint8)
img1_re = cv.dilate(img1_th,se)
img2_re = cv.dilate(img2_th,se)
```

```python
def fillregion(img, x1,x2):
    
    Ac = (255-img)//255 # A complemento

    X = np.zeros(img.shape) # para guardar el relleno
    X[x1,x2] = 1 # pongo el punto inicial de la iteración

    B = cv.getStructuringElement(cv.MORPH_CROSS,(3,3)) #Elemento Estructurante

    X1 = cv.dilate(X,B)*Ac #X0

    while(np.sum(X1)!= np.sum(X)):
        X = X1
        X1 = cv.dilate(X,B)*Ac
    
    X_final = X1*255+img
    return X1, X_final
```

```python
re1, img1_filled = fillregion(img1_re, 130,292)
re2, img2_filled = fillregion(img2_re, 53,140 )
```

```python
plt.figure(figsize=(10,5))
plt.subplot(231); plt.imshow(img1_re, cmap='gray'); plt.title("Imagen 1"); plt.axis("off")
plt.subplot(234); plt.imshow(img2_re, cmap='gray'); plt.title("Imagen 2"); plt.axis("off")
plt.subplot(232); plt.imshow(re1, cmap='gray'); plt.title("Imagen 1 | Región a llenar"); plt.axis("off")
plt.subplot(235); plt.imshow(re2, cmap='gray'); plt.title("Imagen 2 | Región a llenar"); plt.axis("off")
plt.subplot(233); plt.imshow(img1_filled, cmap='gray'); plt.title("Imagen 1 final"); plt.axis("off")
plt.subplot(236); plt.imshow(img2_filled, cmap='gray'); plt.title("Imagen 2 final"); plt.axis("off")
```

# Adelgazamiento


```python
ret, img1_th = cv.threshold(img1, 50, 255, cv.THRESH_BINARY_INV)
ret, img2_th = cv.threshold(img2, 80, 255, cv.THRESH_BINARY_INV)

se1 = cv.getStructuringElement(cv.MORPH_CROSS, (5,5))
se2 = cv.getStructuringElement(cv.MORPH_RECT,(3,5))

img1_th = cv.dilate(img1_th, se1, iterations = 2)
img2_th = cv.dilate(img2_th, se2, iterations = 2)

#Thinning
img1_thin = morph.thin(img1_th,20)
img2_thin = morph.thin(img2_th,20)
img1_sk = morph.skeletonize(img1_th//255)
img2_sk = morph.skeletonize(img2_th//255)

plt.figure(figsize=(12,5))
plt.subplot(231); plt.imshow(img1_th, cmap='gray'); plt.title("Imagen 1"); plt.axis("off")
plt.subplot(234); plt.imshow(img2_th, cmap='gray'); plt.title("Imagen 2"); plt.axis("off")
plt.subplot(232); plt.imshow(img1_thin, cmap='gray'); plt.title("Imagen 1 Adelgazada | 20 iterations"); plt.axis("off")
plt.subplot(235); plt.imshow(img2_thin, cmap='gray'); plt.title("Imagen 2 Adelgazada | 20 iterations"); plt.axis("off")
plt.subplot(233); plt.imshow(img1_sk, cmap='gray'); plt.title("Imagen 1 | Esqueleto"); plt.axis("off")
plt.subplot(236); plt.imshow(img2_sk, cmap='gray'); plt.title("Imagen 2 | Esqueleto"); plt.axis("off")
```
