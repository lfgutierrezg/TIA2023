## Dilatación

```python
img1 = cv2.imread('human detection dataset/1/208.png',0)[50:450,250:400]
img2 = cv2.imread('human detection dataset/1/482.png',0)[80:450,140:300]

ret,img1 = cv2.threshold(img1,90,255,cv2.THRESH_BINARY)
ret,img2 = cv2.threshold(img2,140,255,cv2.THRESH_BINARY)

#Dilation
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
se2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(7,7))

img_dil11 = cv2.dilate(img1,se1, iterations=1)
img_dil12 = cv2.dilate(img1,se2,cv2.BORDER_REPLICATE, iterations=2)
img_dil21 = cv2.dilate(img2,se1,  iterations=1)
img_dil22 = cv2.dilate(img2,se2,cv2.BORDER_REPLICATE, iterations=2)

#Plot
plt.figure(figsize=(15,15))
plt.subplot(331),plt.imshow(se1,cmap='gray')
plt.title('SE1'),plt.xticks([]),plt.yticks([])
plt.subplot(332),plt.imshow(se2,cmap='gray')
plt.title('SE2'),plt.xticks([]),plt.yticks([])
plt.subplot(334),plt.imshow(img1,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(335),plt.imshow(img_dil11,cmap='gray')
plt.title('SE - Cross'),plt.xticks([]),plt.yticks([])
plt.subplot(336),plt.imshow(img_dil12,cmap='gray')
plt.title('SE - Ellipse'),plt.xticks([]),plt.yticks([])
plt.subplot(337),plt.imshow(img2,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(338),plt.imshow(img_dil21,cmap='gray')
plt.title('SE - Cross'),plt.xticks([]),plt.yticks([])
plt.subplot(339),plt.imshow(img_dil22,cmap='gray')
plt.title('SE - Ellipse'),plt.xticks([]),plt.yticks([])

plt.show()
```

<h2>Transformada Hit-or-Miss</h2>


```python

img = cv2.imread('human detection dataset/1/482.png',0)[0:521,0:400]
ret,img_th = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)

#Hit-or-Miss
sel = img[120:350,140:250]

if (sel.shape[0]%2) == 0:
    sel = sel[:sel.shape[0]-1,:]
if (sel.shape[1]%2) == 0:
    sel = sel[:,:sel.shape[1]-1] #Assigns odd dimensions to SE

ret,sel = cv2.threshold(sel,127,255,cv2.THRESH_BINARY_INV)
se = sel//255   #SE must contain 0s and 1s only

se_aux = np.asarray([[1,1],[1,1]])
se = cv2.erode(se,se_aux)        #SE is eroded to allow a more flexible match

img_hom = cv2.morphologyEx(img_th,cv2.MORPH_HITMISS,se)

img_res = np.zeros(img.shape)    #Array to store processed image
h = se.shape[0]//2               #Half the height of SE
w = se.shape[1]//2               #Half the width of SE

print(h,w,sel.shape,img.shape)

counter = 0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img_hom[i,j] == 255:
            if j>=w and i>=h:             # fixes matching bugs at image borders
                img_res[i-h:i+h+1,j-w:j+w+1] = sel
                counter += 1
                
img_res = 255-img_res

print(f"Se han detectado {counter} coincidencias.")
 
#Plot
plt.figure(figsize=(10,5))
plt.subplot(221),plt.title('SE'),plt.imshow(255-se,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.title('Body'),plt.imshow(255-sel,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(223),plt.title('Original'),plt.imshow(img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(224),plt.title('Hit-or-Miss Transformed'),plt.imshow(img_res,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
```

<h2>Llenado de regiones</h2>
<p style='font-size:120%'>$X_k =(X_{k-1}\oplus B)\cap A^c,~~k=1,2,3,...$</p>
<p>Para que esta técnica funcione,<b>TODA región que se quiere rellenar debe tener al menos 1 píxel semilla en la matriz X inicial</b>. Dado que es de tipo recursivo, el proceso se detiene cuando $X_k = X_{k-1}$.</p>

```python
img = cv2.imread('human detection dataset/1/482.png',0)[120:350,140:300]
ret,img_th = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)

#Boundary closing through dilation
se = np.asarray([[1,1,1]],np.uint8)
img = cv2.dilate(img_th,se)

#Hole filling
Ac = (255 - img)//255

X = np.zeros(img.shape)
X[80,65] = 1 #This pixel was identified manually as member of a hole

B = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

X1 = cv2.dilate(X,B)*Ac
while(np.sum(X1) != np.sum(X)):
    X = X1
    X1 = cv2.dilate(X,B)*Ac




    
    
#Plot
plt.figure(1)
plt.subplot(121),plt.title('Original'),plt.imshow(img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.title('Hole filled'),plt.imshow(X1*255+img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()

img=X1*255+img

X = np.zeros(img.shape)
X[63,70] = 1 #This pixel was identified manually as member of a hole

B = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

X1 = cv2.dilate(X,B)*Ac
while(np.sum(X1) != np.sum(X)):
    X = X1
    X1 = cv2.dilate(X,B)*Ac
    
plt.figure(1)
plt.subplot(121),plt.title('Original'),plt.imshow(img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.title('Hole filled'),plt.imshow(X1*255+img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
# print(img[6,10])
```

<h2>Adelgazamiento</h2>
<p style='font-size:120%'>$A\otimes B = A-(A\circledast B)$</p><br>
<a href='https://scikit-image.org/docs/dev/api/skimage.morphology.html#skimage.morphology.thin'>Adelgazamiento con skimage</a>
<br>
<a href='https://scikit-image.org/docs/dev/api/skimage.morphology.html#skimage.morphology.thin'>Esqueletonización con skimage para imágenes 2D</a>

```python
img = cv2.imread('human detection dataset/1/482.png',0)[120:350,140:300]
ret,img_th = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)



#Dilation to ensure connectivity
se = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(7,7))
img_th = cv2.dilate(img_th,se)

#Thinning
img_thin = morph.thin(img_th,20)
# img_thin2 = morph.thin(img_th)
img_thin2 = morph.skeletonize(img_th//255)

#Plot
plt.figure(1)
plt.subplots_adjust(top=2,bottom=0.01,left=0.10,right=2,hspace=0.25,wspace=0.5)

plt.subplot(131),plt.title('Original'),plt.imshow(img_th,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(132),plt.title('Thinned - 20 iterations'),plt.imshow(img_thin,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(133),plt.title('Thinned - Skeleton'),plt.imshow(img_thin2,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
```

<p style='text-align:justified'><b>3.</b> Tome una imagen de su elección y umbralícela para obtenerla a blanco y negro. Luego:<br><br>
a) Aplique una dilatación y luego una erosión sobre ella, y verifique si la erosión deshace los cambios realizados por la dilatación.</p>

```python
img = cv2.imread('human detection dataset/1/62.png',0)[180:400,250:400]
ret,thresh1 = cv2.threshold(img,90,255,cv2.THRESH_BINARY)

#Plot
plt.figure(figsize=(15,15))
plt.subplot(121),plt.imshow(img,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(thresh1,cmap='gray')
plt.title('Umbralizada'),plt.xticks([]),plt.yticks([])
```

```python
#Erosion
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
img_er1 = cv2.erode(thresh1,se1)
#Dilation
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
img_dil1 = cv2.dilate(img_er1,se1)

plt.figure(figsize=(15,15))
plt.subplot(131),plt.imshow(thresh1,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(132),plt.imshow(img_er1,cmap='gray')
plt.title('erosionada'),plt.xticks([]),plt.yticks([])
plt.subplot(133),plt.imshow(img_dil1,cmap='gray')
plt.title('erosionada+dilatada'),plt.xticks([]),plt.yticks([])
```

b) Dilátela usando el mismo SE del inciso anterior, y luego halle el complemento del resultado que obtenga (lo que era blanco, ahora será negro y viceversa).<br>

```python
#Dilation
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
img_dil1 = cv2.dilate(thresh1,se1)
ret,thresh2 = cv2.threshold(img_dil1,127,255,cv2.THRESH_BINARY_INV)

plt.figure(figsize=(15,15))
plt.subplot(131),plt.imshow(thresh1,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(132),plt.imshow(img_dil1,cmap='gray')
plt.title('dilatada'),plt.xticks([]),plt.yticks([])
plt.subplot(133),plt.imshow(thresh2,cmap='gray')
plt.title('complemento dilatada'),plt.xticks([]),plt.yticks([])
```

<p>c) Compare el resultado del inciso b) con el obtenido al aplicar la operación $A^c \ominus \hat{B}$, donde $A$ es su imagen en formato blanco y negro (use el mismo SE del inciso a). Esto se conoce como el <b> principio de dualidad</b>: La dilatación y la erosión no son opuestas, sino que se pueden escribir una en función de la otra.</p>

```python

ret,thresh3 = cv2.threshold(thresh1,127,255,cv2.THRESH_BINARY_INV)
#Erosion
se1 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
img_er1 = cv2.erode(thresh3,se1)


plt.figure(figsize=(15,15))
plt.subplot(131),plt.imshow(thresh1,cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(132),plt.imshow(thresh3,cmap='gray')
plt.title('complemento'),plt.xticks([]),plt.yticks([])
plt.subplot(133),plt.imshow(img_er1,cmap='gray')
plt.title('complemento erosionada'),plt.xticks([]),plt.yticks([])
```

<p style='text-align:justify'><b>7.</b> Ejecute el script revisado arriba para la transformación Hit-or-Miss, pero esta vez utilice como SE la letra <i>t</i> (la ruta es "figs/sehom3.png"). ¿Se reconocen correctamente todas las coincidencias? En caso de que no, ¿Qué se puede hacer para corregir esto? </p>

```python
img = cv2.imread('figs/dilationt1.png',0)
ret,img_th = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)

#Hit-or-Miss
sel = cv2.imread('figs/sehom3.png',0)

if (sel.shape[0]%2) == 0:
    sel = sel[:sel.shape[0]-1,:]
if (sel.shape[1]%2) == 0:
    sel = sel[:,:sel.shape[1]-1] #Assigns odd dimensions to SE

ret,sel = cv2.threshold(sel,127,255,cv2.THRESH_BINARY_INV)
se = sel//255                    #SE must contain 0s and 1s only

se_aux = np.asarray([[1],[1]])
se = cv2.erode(se,se_aux)        #SE is eroded to allow a more flexible match

img_hom = cv2.morphologyEx(img_th,cv2.MORPH_HITMISS,se)

img_res = np.zeros(img.shape)    #Array to store processed image
h = se.shape[0]//2               #Half the height of SE
w = se.shape[1]//2               #Half the width of SE

counter = 0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img_hom[i,j] == 255:
            if j>=w:             # fixes matching bugs at image borders
                img_res[i-h:i+h+1,j-w:j+w+1] = sel
                counter += 1
img_res = 255-img_res

print(f"Se han detectado {counter} coincidencias.")
 
#Plot
plt.figure(figsize=(10,5))
plt.subplot(221),plt.title('SE'),plt.imshow(255-se,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(222),plt.title('Letter t'),plt.imshow(255-sel,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(223),plt.title('Original'),plt.imshow(img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.subplot(224),plt.title('Hit-or-Miss Transformed'),plt.imshow(img_res,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
```

# Analisis

## Dilatación

Se puede observar que conforme se incrementa el tamaño de el elemento estructural se obtiene una mayor dilatación de la zona en tono blanco, tambien se puede ver como segun el numero de iteraciones se obtienen contornos y formas que se asemejan a el elemento estructural.

## Erosión

La erosión de manera similar disminuye las zonas de color blanco o dicho de otra forma 'dilata' el color negro se ve como tambien con las iteraciones adquiere la forma del elemento estructural. Dados los colores de las imagenes se puede observar como si el elemento estructural es demaciado grande se puede perder la forma de la silueta del cuerpo humano.

## Apertura / Cierre

Para nuestro caso en terminos de eliminar posibles discontinuidades que puede encontrarse dentro de la silueta a forma de ruido es la apertura ya que elimina los agujeros de color blanco al interior de la forma humana. Nuevamente se observa como segun el tamaño y geometria del elemento estructural condiciona la forma final obtenida como la similar al rectangulo o a la elipse.

## Transformada Hit or Miss

En esta se pudo evidenciar como la erosión es un paso muy importante en pro de buscar el match entre el SE y las partes de la imagen, por otro lado si el SE es demaciado generico, puede generar match en zonas que no se esperaba que ocurrieran y de esta forma tener coincidencias erroneas.

## Extracción de Frontera

Es absolutamente coherente como a partir de realizar una erosion y posteriormente una resta se obtiene el contorno con un espesor de acuerdo al numero de iteraciones y el tamaño del elemento estructural de la erosión, para las imagenes de nuestro proyecto parece no ser totalmente funcional este tipo de metodologias si las imagenes son binarias ya que por ejemplo el torso de una de las personas no se puede extraer porque desde el inicio en la imagen original este no se distingue.Sin embargo en las zonas donde hay cambio en la coloración la extraccion es absolutamente util y funcional.

## Llenado de regiones

Esta herramienta fui utilizada para llenar dos 'agujeros' de la silueta de una persona en una imagen binaria, en general la metodologia funciona porque despues de dos aplicaciones una para cada una se obtuvo la silueta sin estos dos agujeros. Sin embargo, algo que parece sumamente dificil de automatizar podria ser el hecho de la necesidad de especificar las coordenadas de la semilla. Sin embargo, como se pudo observar en la clase de segmentación con información del histograma es posible seleccionar puntos de semilla para realizar la identificacion de regiones y posteriormente analizar si se clasificaron en su totalidad al contemplar la totalidad del area

## Adelgazamiento

Esta es una de las herramientas que mas puede servirnos en el proceso de identificacion de una persona, ya que en general el cuerpo y sus movimientos generalmente reflejan el esqueleto obtenido, en el que una linea representa la cabeza, dos mas los brazos, dos mas las piernas y una linea central para el torso. Aunque nos parece extraño y algo gracioso la linea horizontal a la altura de la cadera, esta en un proceso de match podria ser omitida ya que no importaria si esta linea estuviera o no alli.Tambien se ve como se pueden obtener resultados parciales de el proceso de esqueletizacion con el adelgazamiento ya que el hecho de tener el torso un poco mas ancho podria faciilitar el proceso de match entre varias poses del cuerpo y sus extremidades.
