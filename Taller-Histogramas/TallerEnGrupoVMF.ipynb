```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgOscura = cv2.imread("pics/oscura.jpg", 0)
imgMedia = cv2.imread("pics/media.jpg", 0)
imgClara = cv2.imread("pics/claro.jpg", 0)


#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.subplot(131),plt.imshow(imgOscura, cmap='gray')
plt.title('Imagen Oscura'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(imgMedia, cmap='gray')
plt.title('Imagen Media'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(imgClara, cmap='gray')
plt.title('Imagen Clara'), plt.xticks([]), plt.yticks([])

plt.show() # Removes additional info displayed
```

**Bins = 2**

```python
bins = 2

histOscura = cv2.normalize(cv2.calcHist([imgOscura], [0], None, [bins], [0,256]),imgOscura[0].size)
histMedia = cv2.normalize(cv2.calcHist([imgMedia], [0], None, [bins], [0,256]),imgMedia[0].size)
histClara = cv2.normalize(cv2.calcHist([imgClara], [0], None, [bins], [0,256]),imgClara[0].size)

plt.figure(figsize=(15,5))

plt.subplot(131).set_title('Oscura'),plt.hist(imgOscura[0], bins, [0,256], color='blue', density=1)
plt.subplot(132).set_title('Media'), plt.hist(imgMedia[0], bins, [0,256], color='blue', density=1)
plt.subplot(133).set_title('Clara'), plt.hist(imgClara[0], bins, [0,256], color='blue', density=1)
"""
plt.hist - Keyword args:
-- image -> 1D array. image to get the histogram from. Use .ravel() to make it 1D
-- bins -> integer. Number of bins - 32
-- cumulative -> bool. Plots cumulative histogram - True
-- density -> bool. Normalizes histogram - True
"""
plt.show()

```

```python
#Comparison 
for i in range(0,4 ):
        
    metric_corr = cv2.compareHist(histOscura,histMedia, i)

    opcion = i

    variable_nombre = ''

    if opcion == 0:
        variable_nombre = 'Correlation'
    elif opcion == 1:
        variable_nombre = 'Chi-square'
    elif opcion == 2:
        variable_nombre = 'Intersection'
    elif opcion == 3:
        variable_nombre = 'Bhattacharyya'
    else:
        print('Opción no válida')

    
    print(f"Correlation (Oscura and Media): {variable_nombre}---{metric_corr}")
```

```python
for i in range(0,4 ):
        
    metric_corr = cv2.compareHist(histOscura,histClara, i)

    opcion = i

    variable_nombre = ''

    if opcion == 0:
        variable_nombre = 'Correlation'
    elif opcion == 1:
        variable_nombre = 'Chi-square'
    elif opcion == 2:
        variable_nombre = 'Intersection'
    elif opcion == 3:
        variable_nombre = 'Bhattacharyya'
    else:
        print('Opción no válida')

    
    print(f"Correlation (Oscura and Clara): {variable_nombre}---{metric_corr}")
```

### Analisis 

- De la figura de los histogramas con 2 bins se puede observar una alta simulitud entre las tres figuras con un leve incremento de los pixeles para el bin de gris mas claro en la fotografia mas clara. Esto puede ser corroborado por cada una de las metricas de comparación ya que para la imagen Oscura y Media los valores de Correlation e intersection son bastante cercanos a 1, compartamiento igual en la comparación entre imagen Oscura y Clara. Los valores de la metrica Chi-square y Bhattacharyya tambien son cercanos a cero. 

- Si observamos en general las pruebas realizadas con los 3 diferentes numeros de bins se puede ver que a medida que se incrementa los posibles valores de gris para agrupar y realizar los histogramas las imagenes se hacen mas diferentes entre si, dado que el valor de gris promedio se va corriendo hacia 255 conforme la imagen se hace mas clara y los histogramas se hacen mas esbeltos al tener mas posibles valores de tono . Ademas las metricas de compración tambien se alejan de sus valores de match de manera mas marcada en la comparación entre imagen Oscura y Clara que entre los valores de Imagen Oscura y Media. 

- Todos los valores de las metricas que se salen de sus rangos normalizados son provocados por los errores de precisión en la normalización y por esto si incrementa el numero de bins incrementa el valor de error asociado sin embargo se entiende que la metrica de Chi-cuadrado si tiene un valor de desajuste ilimitado y entre mas bins se añadan mayor sera el numero para el maximo desajuste. De manera relativa lo que podemos es verificar si la imagen oscura esta mas correlacionada con la imagen media o con la imagen clara.

- La metrica de Correlation entre la imagen Oscura y la Imagen Media se ve como su valor desciende desde 1 hasta 0.64 a medida que aumenta el valor de bins, al aumentar las posibilidades de color gris se notan mas las diferencias entre la iluminacion de las imagenes.

- En los histogramas de las imagenes Media y Clara para 16 y 256 bins se puede ver que la forma de los histogramas es similar pero cambia el valor medio de gris alrededor del cual se forma. Esto permite inferir visualmente que puede estar cambiando la iluminación de la imagen pero no la imagen que se esta analizando sin la necesidad ver la imagen. Esto entendiendo que imagenes diferentes por supuesto pueden tener histogramas iguales y este seria un riesgo.

- La forma diferente entre el histograma de las imagenes Oscura y Clara hace para 16 y 256 bins los valores de las metricas de Correlation y Bhattacharyya sean bastante que representan condiciones similares a pesar del gran cambio en bins.
